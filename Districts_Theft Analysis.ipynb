## 1. Data preparation
<a id='intro'></a>
[Return to: Table of contents](#table_content)

```python
# Libraries
import geopandas as gpd
```

```python
url = "https://www.datos.gov.co/resource/9vha-vh9n.geojson"
thefts = gpd.read_file(url)
thefts
```

```python
# Set the dane_code as index
thefts = thefts.set_index("codigo_dane")
```

```python
# Visualizing the dataframe with the set index.
thefts.head()
```

```python
# Group the dane codes and display the sum of each district "municipio"
thefts_grouped = thefts.groupby("codigo_dane")["municipio"].agg(["count"])
thefts_grouped
```

```python
# Now, it is needed a shp file with the dane code of each district to obtain the spatial information for geopandas
districts = gpd.read_file(r"C:\Users\MEAP\Documents\Python_DataAnalysis\GeoPandas_portfolio\3_Districts_Theft Analysis\COL_Municipio.zip")
districts.head()
```

```python
# Verifying that the "districts" dataframe has all the polygons of Colombia
districts.plot(figsize =(10,8))
```

The dataframes:
* **thefts_grouped** has the "dane_code" and "count" of thefts. On the other hand, **districts** has the "dane_code" and the "polygons data". Therefore, we could combine the data into one dataframe to continue with the analysis.

* In this case we will combine the data with the method "merge" because I am combining attributes.

```python
colombia_merged_df = districts.merge(thefts_grouped, 
                                 left_on = "cod_dane", right_on = "codigo_dane")

colombia_merged_df.head()
```

```python
# Ascending sort of the thefts count
top_5_thefts_areas = colombia_merged_df.sort_values(by = "count", 
                                                    ascending = False)

top_5_thefts_areas.head(5)
```

```python
# Visualizing the districts that have reported thefts
axis = top_5_thefts_areas.plot(figsize = (10,8))
axis
```

## 2. Thefts distribution map
<a id='map_graph'></a>

[Return to: Table of contents](#table_content)

```python
# Plotting the thefts by district over the main layer map of Colombia
axis = districts.plot(figsize = (10,8), color = "#78909c")
graph1 = colombia_merged_df.plot(column = "count", 
                                 ax = axis,
                                 scheme = "QUANTILES", k = 3,
                                 legend = True)

graph1.set_title("Thefts by district in Colombia [South America]", fontsize=16)
graph1.set_xlabel("West - East [abscissa (metre) - Cartesian]")
graph1.set_ylabel("South - North  [ordinate (metre) - Cartesian]")
```

```python
summary = thefts.groupby(["municipio"])["municipio"].count()
```

```python
summary = summary.sort_values(ascending = False).reset_index(name = "nr_thefts").head(10)
```
